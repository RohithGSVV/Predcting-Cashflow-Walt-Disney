```python
import pandas as pd
```

```python
dfb = pd.read_csv('Data/balance-sheet.csv')
dfc = pd.read_csv('Data/cash-flow.csv')
dff = pd.read_csv('Data/financials.csv')
```

```python
dfb.head()
```

```python
dfb = dfb.T
dfc = dfc.T
dff = dff.T
```

```python
dfb.to_excel('balance-sheet.xlsx')
dfc.to_excel('cash-flow.xlsx')
dff.to_excel('financials.xlsx')
```

```python
dfb = pd.read_excel('balance-sheet.xlsx', header=1)
dfc = pd.read_excel('cash-flow.xlsx', header=1)
dff = pd.read_excel('financials.xlsx', header=1)
```

```python
dff.head()
```

```python
dfb.columns = dfb.columns.str.replace('\t', '', regex=True)
dfb.rename(columns={'name': 'Date'}, inplace=True)
dfc.columns = dfc.columns.str.replace('\t', '', regex=True)
dfc.rename(columns={'name': 'Date'}, inplace=True)
dff.columns = dff.columns.str.replace('\t', '', regex=True)
dff.rename(columns={'name': 'Date'}, inplace=True)
```

```python
def data_numeric(df):
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    
    for col in df.columns:
        if col != 'name':
            if df[col].dtype == 'object':
                df[col] = df[col].str.replace(',', '').astype(float)
            else:
                df[col] = pd.to_numeric(df[col], errors='coerce')
    
    return df
```

```python
dfb = data_numeric(dfb)
```

```python
dfc = data_numeric(dfc)
dff = data_numeric(dff)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24, 6))
# Balance Sheet
dfb.plot(x='Date', y=['TotalAssets', 'TotalLiabilitiesNetMinorityInterest'], ax=axes[0], title='Total Assets vs Total Liabilities Over Time')
axes[0].set_ylabel('Amount ($)')
axes[0].set_xlabel('Date')
axes[0].tick_params(axis='x', rotation=45)
# Cash Flow
dfc.plot(x='Date', y=['OperatingCashFlow', 'FreeCashFlow'], ax=axes[1], title='Operating Cash Flow vs Free Cash Flow Over Time')
axes[1].set_ylabel('Amount ($)')
axes[1].set_xlabel('Date')
axes[0].tick_params(axis='x', rotation=45)
# Financials
dff.plot(x='Date', y=['TotalRevenue', 'NetIncome'], ax=axes[2], title='Total Revenue vs Net Income Over Time')
axes[2].set_ylabel('Amount ($)')
axes[2].set_xlabel('Date')
axes[0].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
